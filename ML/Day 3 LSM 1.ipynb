```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/content/LSM1.csv")
```

```python
df
```

```python
x = df.iloc[:,0].values.reshape(5,1)
y = df.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lin = LinearRegression()
```

```python
Lin.fit(x,y)
```

```python
Lin.predict([[2.5]])
```

```python
pred_y = Lin.predict(x)
```

```python
pred_y
```

```python
y
```

```python
plt.scatter(x,y)
plt.plot(x, pred_y, c = "g", marker = "*")
plt.show()
```
