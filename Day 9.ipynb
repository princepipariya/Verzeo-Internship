#Plots

```python
import matplotlib.pyplot as plt
```

```python
x = [1,4,8,10]
y = [2,9,15,19]
plt.plot(x,y, marker = "o", color = "green", linestyle = "dashed",
         markersize = 10, linewidth = 2)
plt.xlabel("X-label")
plt.ylabel("Y- label")
plt.show()
```

```python
x = [1,4,8,10]
y = [2,9,15,19]
plt.scatter(x,y, marker = "o", color = "green", s = 50)
plt.xlabel("X-label")
plt.ylabel("Y- label")
plt.show()
```

#Bargraph

```python
import numpy as np
```

```python
name = ["Chaitanya", "Deepashree", "Sappu", "Aditya", "Dhiren", "Thrinath", "Varshi"]
salary = np.random.randint(10000, 50000, 7)
```

```python
salary
```

```python
color_name = ["r","g","b","k","#3f6662","#6257d9", "#b713ed" ]
plt.bar(name, salary, color = color_name, width= 0.8)
plt.xlabel("Name")
plt.ylabel("Salary")
plt.show()
```
