Functions:
1. User-defined
2. Pre-defined

```python
def greet_user():
  print("Hi there...")
  print("Welcome aboard")
```

```python
print("Start")
greet_user()
print("Stop")
```

```python
def greet_user(name):       #Parameter
  print(f"Hi there {name}")
  print("Welcome aboard")
```

```python
greet_user("Liqzan")        #Argument
```

```python
def greet_user(first, last):       #Parameter
  print(f"Hi there {first} {last}")
  print("Welcome aboard")
```

```python
greet_user("Liqzan", "Manna")
```

```python
greet_user(last = "Manna", first= "Liqzan")
```
